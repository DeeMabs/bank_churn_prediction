```python
#import required libraries needed
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score,classification_report

#importing the CSV
Churning=pd.read_csv(r"C:\Users\user\Desktop\Part 1.2\FE 1.2\Lab II Financial Analytics and Big Data in Finance\Bank Customer Churn Prediction.csv")

#printing the head 
Churning.head()

#Looking for missing values
Churning.isnull()
#Check if there is no missing values
Churning.isnull().sum()

#Dropping the Customerid and country because they are not necessary
Churning=Churning.drop(columns=['country','customer_id'])

#changing values to numeric values

Churning['gender']=pd.Categorical(Churning['gender']).codes


x=Churning.drop(['churn'],axis=1).values
y=Churning['churn'].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
#intialization of models
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier(),
    "BC": BaggingClassifier(),
}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    print(f"{name}: Accuracy = {accuracy:.5f}")
    classification = classification_report(y_test,y_pred)
    print(f"{name}\n{classification}\n")

    
```
